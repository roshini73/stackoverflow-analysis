### e) User's question to answer ratio 


#### i) Question Score

```
%%bigquery --project $project_id q2fa1

# YOUR QUERY HERE

SELECT q2a, AVG(qScore) AS avgqScore
FROM (
SELECT COUNT(*)/aCount AS q2a, q.score AS qScore
FROM `bigquery-public-data.stackoverflow.posts_questions`q
JOIN (
SELECT owner_user_id AS userID, COUNT(*) AS aCount
FROM `bigquery-public-data.stackoverflow.posts_answers`
GROUP BY owner_user_id
) a
ON a.userID = q.owner_user_id
GROUP BY q.owner_user_id, aCount, q.score
)
GROUP BY q2a
```

```
%%bigquery --project $project_id q2fa2

# YOUR QUERY HERE
SELECT q2a, qScore, countS AS numOfInstances,COUNT(*)/countS AS freq
FROM (
SELECT 
CASE WHEN ROUND(COUNT(*)/aCount, 2) < 0.25 THEN '<0.25'
WHEN ROUND(COUNT(*)/aCount, 2) < 0.5 THEN '<0.5'
WHEN ROUND(COUNT(*)/aCount, 2) < 0.75 THEN '<0.75'
ELSE '<1' END AS q2a, q.score AS qScore
FROM `bigquery-public-data.stackoverflow.posts_questions`q
JOIN (
SELECT owner_user_id AS userID, COUNT(*) AS aCount
FROM `bigquery-public-data.stackoverflow.posts_answers`
GROUP BY owner_user_id
) a
ON a.userID = q.owner_user_id
GROUP BY q.owner_user_id, aCount, qScore
) r
JOIN (
SELECT score, COUNT(*) AS countS
FROM `bigquery-public-data.stackoverflow.posts_questions` 
GROUP BY score
) s
ON r.qScore = s.score
GROUP BY q2a, qScore, countS
```

```
%%bigquery --project $project_id q2fa3

# YOUR QUERY HERE
SELECT q2a, qScore, countS AS numOfInstances,COUNT(*)/countS AS freq
FROM (
SELECT 
CASE WHEN ROUND(COUNT(*)/aCount, 2) < 0.25 THEN '<0.25'
WHEN ROUND(COUNT(*)/aCount, 2) < 0.5 THEN '<0.5'
WHEN ROUND(COUNT(*)/aCount, 2) < 0.75 THEN '<0.75'
ELSE '<1' END AS q2a, q.score AS qScore
FROM `bigquery-public-data.stackoverflow.posts_questions`q
JOIN (
SELECT owner_user_id AS userID, COUNT(*) AS aCount
FROM `bigquery-public-data.stackoverflow.posts_answers`
GROUP BY owner_user_id
) a
ON a.userID = q.owner_user_id
GROUP BY q.owner_user_id, aCount, qScore
) r
JOIN (
SELECT score, COUNT(*) AS countS
FROM `bigquery-public-data.stackoverflow.posts_questions` 
GROUP BY score
) s
ON r.qScore = s.score
WHERE r.qScore < 5000
GROUP BY q2a, qScore, countS
```

```
# YOUR PLOT CODE HERE
import altair as alt 
alt.data_transformers.enable('default', max_rows=None)

alt.Chart(q2fa1, title="Average Question Score vs. User Question to Answer Ratio").mark_bar().encode(
    x=alt.X('q2a:Q', scale=alt.Scale(type='log'), axis=alt.Axis(title='Question to Answer ratio')),
    y=alt.Y('avgqScore', axis=alt.Axis(title='Average Question Score')),
)
```

```
# YOUR PLOT CODE HERE
import altair as alt
alt.data_transformers.enable('default', max_rows=None)

alt.Chart(q2fa2, title="Proportions of Question to Answer Ratios vs. Question Score Range").mark_bar().encode(
    x=alt.X('sum(freq)', stack="normalize", axis=alt.Axis(title='Proportions of Question to Answer Ratios')),
    y=alt.Y('qScore', bin=True, axis=alt.Axis(title='Question Score Range')),
    color=alt.Color('q2a', sort=('<0.25', '<0.5', '<0.75', '<1')),
    tooltip=['q2a', 'freq', 'qScore', 'numOfInstances']
).interactive()
```

```
# YOUR PLOT CODE HERE
import altair as alt
alt.data_transformers.enable('default', max_rows=None)

alt.Chart(q2fa3, title="Proportions of Question to Answer Ratios vs. Question Score Range for Score < 5000").mark_bar().encode(
    x=alt.X('sum(freq)', stack="normalize", axis=alt.Axis(title='Proportions of Question to Answer Ratios')),
    y=alt.Y('qScore', bin=True, axis=alt.Axis(title='Question Score Range')),
    color=alt.Color('q2a', sort=('<0.25', '<0.5', '<0.75', '<1')),
    tooltip=['q2a', 'freq', 'qScore', 'numOfInstances']
).interactive()
```

#### ii) Answer Score

```
%%bigquery --project $project_id q2fb1

# YOUR QUERY HERE

SELECT q2a, AVG(aScore) AS avgaScore
FROM (
SELECT qCount/COUNT(*) AS q2a, a.score AS aScore
FROM `bigquery-public-data.stackoverflow.posts_answers`a
JOIN (
SELECT owner_user_id AS userID, COUNT(*) AS qCount
FROM `bigquery-public-data.stackoverflow.posts_questions`
GROUP BY owner_user_id
) q
ON q.userID = a.owner_user_id
GROUP BY a.owner_user_id, qCount, a.score
)
GROUP BY q2a
```

```
%%bigquery --project $project_id q2fb2

# YOUR QUERY HERE

SELECT q2a, aScore, countS AS numOfInstances, COUNT(*)/countS AS freq
FROM (
SELECT 
CASE WHEN ROUND(1/COUNT(*)/qCount, 2) < 0.25 THEN '<0.25'
WHEN ROUND(1/COUNT(*)/qCount, 2) < 0.5 THEN '<0.5'
WHEN ROUND(1/COUNT(*)/qCount, 2) < 0.75 THEN '<0.75'
ELSE '<1' END AS q2a, a.score AS aScore
FROM `bigquery-public-data.stackoverflow.posts_answers`a
JOIN (
SELECT owner_user_id AS userID, COUNT(*) AS qCount
FROM `bigquery-public-data.stackoverflow.posts_questions`
GROUP BY owner_user_id
) q
ON q.userID = a.owner_user_id
GROUP BY a.owner_user_id, qCount, a.score
) r
JOIN (
SELECT score, COUNT(*) AS countS
FROM `bigquery-public-data.stackoverflow.posts_answers` 
GROUP BY score
) s
ON r.aScore = s.score
GROUP BY q2a, aScore, countS
```
